```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
url = '/Users/amirrezakamkar/Downloads/amz_uk_price_prediction_dataset.csv'
df = pd.read_csv(url)
```

```python
#1:
crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])
crosstab_result.columns = crosstab_result.columns.astype(int) 
crosstab_result['proportion_best_sellers'] = crosstab_result[1] / (crosstab_result[1] + crosstab_result[0])

sorted_crosstab = crosstab_result.sort_values(by='proportion_best_sellers', ascending=False)

print(sorted_crosstab)
```

```python
from scipy.stats import chi2_contingency
chi2_statistic, chi2_p_value, _, _ = chi2_contingency(sorted_crosstab)
chi2_statistic, chi2_p_value
```

```python
crosstab_result.drop(columns=['proportion_best_sellers'], inplace=True)
```

```python
from scipy.stats.contingency import association

association(crosstab_result, method="cramer")
```

```python
crosstab_result.plot(kind="bar", stacked=True)
```

```python
#2:
def remove_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    filtered_data = data[(data >= lower_bound) & (data <= upper_bound)]
    
    return filtered_data

data = df['price']  
filtered_data = remove_outliers(data)
print(filtered_data)
```
